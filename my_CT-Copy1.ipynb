# Data
Load the Iris dataset from *Scikit-learn* library and convert it to *Pandas* dataframe

```python
data = datasets.load_iris()
data
```

```python
#make pandas.DataFrame with iris data
df=pd.DataFrame(data=data['data'], columns=data['feature_names'])
df['class']=data['target']
df.head(5)
```

## Visual
Display data with suitable plot

```python
#visual 2D 
x='petal width (cm)'
y='petal length (cm)'

plt.figure(figsize=(4,4))
plt.scatter(df[x],df[y],c=df['class'])
plt.xlabel(x)
plt.ylabel(y)
plt.title('Iris')
```
